# 1D DISCREATE FOURIER TRANSFORM

```python
import numpy as np
```

```python
from numpy import fft
```

```python
import matplotlib.pyplot as plt
```

```python
# define number of samples, sample frequency, time
num_of_samples = 400
sample_frequency = 1/200
times = np.linspace(0,num_of_samples*sample_frequency,num_of_samples) # num_of_samples is it takes aroud 0 to 400 values
```

```python
simple_signal=np.sin(20*np.pi*times)
```

```python
plt.plot(times,simple_signal)
plt.show() # we show as 1 carrier, 2 carrier like that 
```

```python
# use fft.rfft() 
fourier_frequency = fft.rfft(simple_signal) # rfft is real fourier transform, fft is fourier transform
```

```python
fourier_frequency.shape # total 201 entries we have
```

```python
#use fft.rfftfreq() that is creating x values
frequency_domain = fft.rfftfreq(num_of_samples,sample_frequency)
```

```python
plt.plot(frequency_domain,fourier_frequency.real) # shows real part
```

```python
plt.plot(frequency_domain,fourier_frequency.imag) #shows imaginary part
```

```python
# add two signals and plot
two_signals = np.sin(100*np.pi*times)+np.sin(40*np.pi*times)
```

```python
plt.plot(times,two_signals)
```

```python
fourier_transform2 = fft.rfft(two_signals)
```

```python
fourier_transform2.shape
```

```python
plt.plot(frequency_domain,fourier_transform2.real)
```

```python
plt.plot(frequency_domain,fourier_transform2.imag)
```

```python
# blue is real part, orange is imaginary part
plt.plot(frequency_domain,fourier_transform2.real)
plt.plot(frequency_domain,fourier_transform2.imag)
```

```python
# set 40 to 0
fourier_transform2[frequency_domain>40]=0
```

```python
plt.plot(frequency_domain,fourier_transform2.real)
plt.plot(frequency_domain,fourier_transform2.imag)
```

```python
# irfft is inverse real fourier transform
low_frequency =fft.irfft(fourier_transform2)
```

```python
# two_signals = np.sin(100*np.pi*times)+np.sin(40*np.pi*times) = from this we use np.sin(40*np.pi*times) this part only
plt.plot(times,low_frequency) 
```

# SMOOTHING A SIGNAL

```python
times = np.linspace(0,2,600) # we need 600 values between 0 and 2
```

```python
signal = np.exp(-np.cos(30*np.pi*times)**2)*(np.cos(10*np.pi*times)+np.sin(100*np.pi*times))
```

```python
plt.plot(times,signal)
```

```python
fft.rfft(signal)
```

```python
fft.rfft(signal.shape)
```

```python
f_t=fft.rfft(signal)
```

```python
frequency_dom = fft.rfftfreq(600,1/300) # initialy we take 600 values,"times = np.linspace(0,2,600)" is 2/600 is 1/300
```

```python
plt.plot(frequency_dom,f_t)
```

```python
plt.plot(frequency_dom,f_t.real)
```

```python
plt.plot(frequency_dom,f_t.imag)
```

```python
plt.plot(frequency_dom,f_t.real)
plt.plot(frequency_dom,f_t.imag)
```

```python
f_t[frequency_dom>40]=0 # set value frequency_dom >40 is 0
```

```python
plt.plot(frequency_dom,f_t.real)
plt.plot(frequency_dom,f_t.imag)
```

```python
#smooting is process removing noise that is removing unwanted up and downs
smooth = fft.irfft(f_t)
```

```python
plt.plot(times,smooth) # see difference between plt.plot(times,signal) and plt.plot(times,smooth)
```

```python
plt.plot(times,signal)
```

# 2D FOURIER TRANSFORM

```python
from skimage.data import camera
```

```python
camera = camera()
```

```python
plt.imshow(camera)
```

```python
# to show image in black and white
plt.imshow(camera,cmap=plt.cm.gray)
```

```python
# take fourier transform for the image
F_camera = fft.rfft(camera)
```

```python
F_camera
```

```python
F_camera.shape
```

```python
#see how fourier transfom looks like for camera image
log = np.log(1+np.abs(F_camera))/np.log(1+np.abs(F_camera).max()) # take'1+' is used to should not get negative values, 
# max() is used to get values between 0 and 1
```

```python
plt.imshow(log,cmap=plt.cm.gray)
```

```python
np.count_nonzero(F_camera)  # counting non-zero values
```

```python
F_camera[np.abs(F_camera) < 1000] = 0
```

```python
log = np.log(1+np.abs(F_camera))/np.log(1+np.abs(F_camera).max())
plt.imshow(log,cmap=plt.cm.gray) # balck spaces are 0
```

```python
np.count_nonzero(F_camera) # see difference between that iamge count here it is reduced
```

```python
# take inverse real fourier transform
compressed = fft.irfft(F_camera)
```

```python
plt.imshow(compressed, cmap=plt.cm.gray) 
# see difference between original iamge and compressed image their won't be lot of difference
```
